Panathinaikos Euroleague 20-21 Stats Analysis

```python
import glob
import pandas as pd
import numpy as np
import seaborn as sns
import time
%matplotlib inline
```

```python
data = pd.read_csv("C:\\Users\\dilevent\\Desktop\\PAO_EL_20_21.csv")
```

```python
PLOT_TOP = 7
data.head(5)
```

```python
plot = data.sort_values(['MP'], ascending=False).plot.pie(y='MP', labels=data['Player'], title = "Players by minutes per game", autopct='%1.1f%%', legend = False, shadow=False, figsize=(16,8))
```

```python
data[data['3PA']>1].sort_values(['3P%'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='3P%', rot=0, title = "Top-{} 3 point shooters (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data[data['PTS']>0.5].sort_values(['PTS'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='PTS', rot=0, title = "Top-{} scorers (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data.sort_values(['TRB'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='TRB', rot=0, title = "Top-{} rebounders (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data.sort_values(['AST'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='AST', rot=0, title = "Top-{} in assists (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data['AST_TO_TOV'] = data['AST'] / data['TOV']
```

```python
data.sort_values(['AST_TO_TOV'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='AST_TO_TOV', rot=0, title = "Top-{} in assist to turnover (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data.sort_values(['TOV'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='TOV', rot=0, title = "Top-{} in turnovers (avg)".format(PLOT_TOP), figsize=(14,4))
```

```python
data.sort_values(['STL'], ascending=False).head(PLOT_TOP).plot.bar(x='Player', y='STL', rot=0, title = "Top-{} in steals (avg)".format(PLOT_TOP), figsize=(14,4))
```
